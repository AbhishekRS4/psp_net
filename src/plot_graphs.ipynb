```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics = np.load('model_120/metrics.npy').item()
```

```python
x_values = np.arange(1, len(metrics['train_loss']) + 1)

plt.figure(figsize = (12, 8))
plt.title('loss vs epoch')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.plot(x_values, metrics['train_loss'], 'b--', label = 'psp-net training loss')
plt.plot(x_values, metrics['valid_loss'], 'r--', label = 'psp-net validation loss')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.title('accuracy vs epoch')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.plot(x_values, metrics['valid_acc'], 'y--', label = 'psp-net validation accuracy')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.title('mean IoU vs epoch')
plt.xlabel('epoch')
plt.ylabel('mean IoU')

plt.plot(x_values, metrics['valid_iou'], 'g--', label = 'psp-net validation mean IoU')

plt.legend()
plt.show()
```
